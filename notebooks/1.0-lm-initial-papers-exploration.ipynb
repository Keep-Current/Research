```python
from gensim.corpora.dictionary import Dictionary
```

```python
corpus = df.paper_text.tolist()
```

```python
nlp = spacy.load('en')
```

```python
test_doc = nlp(corpus[0].replace('\n',' '))
test_doc
```

```python
corpus[0]
```

```python
tokenized_corpus = [[w.text for w in nlp(doc.replace('\n',' '))] for doc in corpus[:100]]
```

```python
tokenized_corpus = [[w for w in t if len(w) > 3] for t in tokenized_corpus]
tokenized_corpus
```

```python
corpus_dict = Dictionary(tokenized_corpus)
common_corpus = [corpus_dict.doc2bow(text) for text in tokenized_corpus]
corpus_dict.token2id
```

```python
from gensim.models import LdaModel
lda = LdaModel(common_corpus, num_topics=10)
```

```python
unseen_text = [[w.text for w in nlp(doc.replace('\n',' '))] for doc in corpus[200:203]]
common_corpus = [corpus_dict.doc2bow(text) for text in unseen_text]
common_corpus = [[w for w in t if len(w)>3] for t in common_corpus]
```

```python
vector = lda[common_corpus[2]]
```
